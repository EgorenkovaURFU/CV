```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

Load an image

```python
img = cv2.imread('data/Lenna.png')
```

For computer it looks like:

```python
print(img)
```

And now, using matplotlib draw image

```python
plt.imshow(img)
plt.show()
```

This image is with blue color. The reason is the image was read as BGR, not as RGB, but we can correct it.

```python
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
plt.show()
```

# Cropping images

```python
img.shape
```

```python
cropped_img = img[80:400, 100:380]
plt.imshow(cropped_img)
plt.show()
```

# Change size

```python
resize_img = cv2.resize(img, (600, 400), interpolation=cv2.INTER_AREA)
plt.imshow(resize_img)
plt.show()
```

# Draw rectangle

```python
output = img.copy()
cv2.rectangle(output, (180, 400), (400, 200), (0, 255, 255), 2)
plt.imshow(output)
plt.show()
```

# Text on image

```python
output = img.copy()
cv2.putText(output, 'Girl', (200, 480), cv2.FONT_HERSHEY_SIMPLEX, 3, (30, 105, 210), 4)
plt.imshow(output)
plt.show()
```

# Gray gradient and using thresholder

```python
gray_img = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
_, thresholded_img = cv2.threshold(gray_img, 127, 255, 0)
plt.subplot(121)
plt.imshow(gray_img, cmap='gray')
plt.subplot(122)
plt.imshow(thresholded_img, cmap='gray')
plt.show()
```

# Filters

```python
noise = np.random.randint(0, 255, (2048, 2048, 3)).astype(np.uint8)
noisy_img = cv2.addWeighted(img, 0.7, noise, 0.3, 2.2) # you can get mix of images using this way
```

```python
gblurred_img = cv2.GaussianBlur(noisy_img,  (21, 21), 0)
mblurred_img = cv2.medianBlur(noisy_img, 21)
plt.figure(figsize=(15, 5))
plt.subplot(131)
plt.title('Noisy Lena')
plt.imshow(noisy_img)
plt.subplot(132)
plt.title('Gaussian smoothing')
plt.imshow(gblurred_img)
plt.subplot(133)
plt.title('Median smoothing')
plt.imshow(mblurred_img)
plt.show()
```
